```python
from sklearn import datasets
import pandas as pd
iris = datasets.load_iris()
```

```python
X=pd.DataFrame(iris.data)
```

```python
y=pd.DataFrame(iris.target)
```

```python
from sklearn.preprocessing import LabelEncoder
encoder =  LabelEncoder()
y1 = encoder.fit_transform(y)

Y = pd.get_dummies(y1).values
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X
```

```python
import keras

from keras.models import Sequential
from keras.layers import Dense

classifier=Sequential()
```

```python
classifier.add(Dense(8, input_dim=4, activation='relu'))

classifier.add(Dense(3, activation='softmax'))

```

```python
classifier.compile(optimizer='adam', loss='binary_crossentropy', metrics = ['accuracy'])
classifier.fit(X_train, y_train, batch_size = 10, epochs = 100)

```

```python
ypred=classifier.predict(X_test)
```

```python
import numpy as np
ypred_max=np.argmax(ypred, axis=1)
```

```python
ypred_max
```

```python
y_test_max=np.argmax(y_test, axis=1)
```

```python
y_test_max
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_test_max,ypred_max))
print(confusion_matrix(y_test_max,ypred_max))
```

```python
def baseline_model():
    classifier=Sequential()
    classifier.add(Dense(8, input_dim=4, activation='relu'))
    classifier.add(Dense(3, activation='softmax'))
    classifier.compile(optimizer='adam', loss='binary_crossentropy', metrics = ['accuracy'])
    return classifier
```

```python
from keras.wrappers.scikit_learn import KerasClassifier
```

```python
estimator=KerasClassifier(build_fn=baseline_model, epochs=200, batch_size=10)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
```

```python
kfold=KFold(n_splits=10, shuffle=True, random_state=10)
results=cross_val_score(estimator, X_train, y_train, cv=kfold)
```

```python
print("Baseline: %.2f%% (%.2f%%)" % (results.mean()*100, results.std()*100))
```
